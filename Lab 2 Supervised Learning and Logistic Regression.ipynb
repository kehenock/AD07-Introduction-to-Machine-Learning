```python
# Step 1: Explore the Dataset

# 1️⃣ Check the shape of the dataset
print("Dataset shape (rows, columns):", data.shape)

# 2️⃣ Preview first 5 rows
print("\nFirst 5 rows of the dataset:\n")
print(data.head())

# 3️⃣ Quick statistical summary
print("\nSummary statistics:\n")
print(data.describe())

# 4️⃣ Check column names
print("\nColumn names:\n")
print(data.columns.tolist())
```

```python
# Step 2: Clean the Data

# 1️⃣ Preview the dataset
print("First 5 rows of the dataset:\n")
print(data.head())

# 2️⃣ Check for missing values
print("\nMissing values per column:\n")
print(data.isnull().sum())

# 3️⃣ Handle zeros in columns that cannot be zero
# In the Pima dataset, some features like Glucose, BloodPressure, BMI, SkinThickness, and Insulin should not be zero
cols_with_zeros = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

# Replace zeros with NaN
data[cols_with_zeros] = data[cols_with_zeros].replace(0, np.nan)

# Check again for missing values after replacement
print("\nMissing values after replacing zeros with NaN:\n")
print(data.isnull().sum())

# Optional: Impute missing values with column mean
data.fillna(data.mean(), inplace=True)
print("\nMissing values after imputation:\n")
print(data.isnull().sum())
```

```python
# Step 3: Encode the Target Variable

# The target variable 'Outcome' is already binary:
# 0 = No diabetes
# 1 = Diabetes
# No further encoding is required

y = data["Outcome"]
print("Unique values in target variable:", y.unique())
```

```python
# Step 4: Define Features and Scale the Data

# 1️⃣ Define features (X) and target (y)
features = ["Glucose", "BMI", "Insulin", "Age"]
X = data[features]
y = data["Outcome"]  # Target: 1 = diabetes, 0 = no diabetes

# 2️⃣ Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Optional: convert back to DataFrame for easy viewing
X_scaled = pd.DataFrame(X_scaled, columns=features)
print("Scaled feature preview:\n", X_scaled.head())

# 3️⃣ Split the dataset into training (80%) and testing (20%) sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
# Step 5: Build and Train the Logistic Regression Model

# 1️⃣ Instantiate the logistic regression model
model = LogisticRegression(random_state=42)

# 2️⃣ Train the model on the training set
model.fit(X_train, y_train)

# 3️⃣ Make predictions

# Predict probabilities for class 1 (diabetes)
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Predict class labels (0 = no diabetes, 1 = diabetes)
y_pred = model.predict(X_test)

# Show first 10 predicted probabilities and classes
print("Predicted probabilities (first 10):", y_pred_proba[:10])
print("Predicted classes (first 10):", y_pred[:10])
```
